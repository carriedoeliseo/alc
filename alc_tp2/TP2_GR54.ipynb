### Consigna 5: Aplicar Minimos Cuadrados sobre los precios de cada nutriente (Hc, Proteinas, Grasas).

#### Obtencin de los precios de cada nutriente.

Para esta consigna, primero seleccionamos los macronutrientes de los alimentos de consumidores libres por gramo.

```python
# Nos fijamos Hc, Proteinas y Grasas de los precios de consumidores libres
consumidores_libres_macros = separar_macros(consumidores_libres_nut)
consumidores_libres_macros
```

Luego calculamos el precio por gramo de cada alimento

```python
# Calculamos el precio por gramo de cada alimento de consumidores libres
consumidores_libres_pgramo = proporciones_por_gramo(consumidores_libres)

# Nos deshacemos de la berenjema pues no est en la tabla nutricional
consumidores_libres_pgramo = consumidores_libres_pgramo.drop(10,axis=0)
consumidores_libres_pgramo = consumidores_libres_pgramo.reset_index(drop=True)
consumidores_libres_pgramo
```

Y luego el precio por gramo de cada macronutriente de cada alimento.

Para esto, realizamos el siguiente calculo:
$$ 1~gr~Alimento --- X~gr~Nutriente $$
$$ G~gr~Alimento --- 1~gr~Nutriente $$

Queremos calcular $G$ para saber cuantos gramos de alimento necesitamos para conseguir 1 gramo de nutriente $\Rightarrow G = 1/X$

Luego para saber cuanto nos salen $G$ gramos de alimento, unicamente debemos multiplicarlo por su precio $\Rightarrow$ Si $P$ es precio del alimento, el precio que necesitamos para llegar al gramo de nutriente es $G*P = P/X$

En conclusin, debemos dividir el precio por gramo del alimento por la cantidad de nutriente que tiene en un gramo.

```python
# Calculamos el precio de cada macronutriente de cada alimento por gramo
precio_hc, precio_proteina, precio_grasa = tablas_macros(consumidores_libres_pgramo, consumidores_libres_macros)
```
