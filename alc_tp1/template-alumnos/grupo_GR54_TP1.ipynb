### Pregunta 1 - ¿Por que la matriz A definida en (4) es equivalente a $p \ \mathbf{W} \ \mathbf{D} + e \ z^t$ ? Justificar.

A $A$ la podemos reescribir como la suma de dos matrices $A_1$ y $A_2$ donde:

$A = A_1 + A_2$

$a_{1~ij} = \left\lbrace\begin{array}{c}&\ (p~w_{ij})/c_{j} \hspace{1cm} &\ si ~ c_{j} \neq 0 \\&\ 0 \hspace{1cm} &\ si ~ c_{j} = 0  \end{array}\right.$

$a_{2~ij} = \left\lbrace\begin{array}{c}&\ (1-p)/n &\ si ~ c_{j} \neq 0 \\&\ 1/n \hspace{1cm} &\ si ~ c_{j} = 0  \end{array}\right.$

Notar que $A_{2}$ solo varía entre columnas, las filas son siempre iguales; de hecho, es igual a $e~z^{t}~$:

$e~z^{t} = \begin{pmatrix}1 \\ \vdots \\ 1 \end{pmatrix} ~  \begin{pmatrix} z_{0} & ... & z_{j} \end{pmatrix} = \begin{pmatrix} z_{0} & ... & z_{j} \\ z_{0} & ... & z_{j} \\ \vdots & \ddots & \vdots \\ z_{0} & ... & z_{j} \end{pmatrix} = A_2 \hspace{0.5cm}$ con $\hspace{0.5cm} z_{j} = \left\lbrace\begin{array}{c}&\ (1-p)/n &\ si ~ c_{j} \neq 0 \\&\ 1/n \hspace{1cm} &\ si ~ c_{j} = 0  \end{array}\right.$

Luego como $e~z^{t} = A_2~$ nos quedaría ver que $p~W~D = A_{1}~$. Calculemos $p~W~D~$:

Como $W$ es una matriz de unos y ceros, y $D$ una matriz diagonal, su producto matricial $R$ será:

$r_{ij} = w_{ij}*d_{jj} \hspace{0.5cm}$ con $\hspace{0.5cm} d_{jj}$ = $\left\lbrace\begin{array}{c}&\ 1/c_{j} \hspace{1cm} &\ si ~ c_{j} \neq 0 \\&\ 0 \hspace{1cm} &\ si ~ c_{j} = 0 \end{array}\right. \hspace{0.5cm} \Rightarrow \hspace{0.5cm} r_{ij} = \left\lbrace\begin{array}{c}&\ w_{ij}/c_{j} \hspace{1cm} &\ si ~ c_{j} \neq 0 \\&\ 0 \hspace{1cm} &\ si ~ c_{j} = 0 \end{array}\right.$ 

Por último multiplicamos a la matriz $R$ por el escalar $p$ y nos queda:

$r_{ij} = \left\lbrace\begin{array}{c}&\ (p~w_{ij})/c_{j} \hspace{1cm} &\ si ~ c_{j} \neq 0 \\&\ 0 \hspace{1cm} &\ si ~ c_{j} = 0 \end{array}\right. \hspace{0.5cm} = A_{1}$

Finalmente como probamos que $A_{1} = p~W~D~$, $~e~z^{t} = A_2~$ y $~A = A_1 + A_2$

$\therefore ~A = p~W~D + e~z^{t}$



### Pregunta 2 -  ¿Cómo se garantiza existencia de la factorización $LU$? ¿ La matriz $(\mathbf{I} - p \ \mathbf{W} \ \mathbf{D})$ está bien condicionada? ¿Cómo influye el valor de $p$?

$\bullet ~$ ¿Cómo se garantiza existencia de la factorización $LU$?


Primero empecemos demostrando que $A$ tiene factorización $L~U $ si y sólo si $ A^{T}~$ también tiene factorización $L~U$

$\Rightarrow )$ Suponemos que $A$ tiene factorización $L~U \Rightarrow A^{T} = U^{T}L^{T}$

Donde $U^{T}$ es una matriz triangular inferior con $~u_{ii} \neq 0~$ para cualquier elemento de su diagonal; y $L^{T}$ es una matriz triangular superior con $l_{ii} = 1$ para cualquier elemento de su diagonal.

Como para toda matriz triangular $~A \in \mathbb{R}^{n \times n}$, $~det(A) = \prod\limits_{i = 1}^{n}{a_{ii}} ~ $ y $~U^{T}$, $L^{T}~$ lo son y tienen todos los elementos de su diagonal distintos de 0, podemos decir que son invertibles y, más aún, podemos decir que todas sus submatrices principales son invertibles ya que todas cumplen las propiedades anteriores.

Luego, como producto de matrices invertibles es invertible ya que $det(AB) = det(A) * det(B)$ podemos decir que $A^{T}$ es invertible. La misma lógica se aplica para el producto de sus submatrices principales, ya que si lo pensamos como producto matricial por bloques con:

$U^{T}~L^{T}$ = $A^{T} \hspace{0.3cm} = \hspace{0.3cm}
\begin{pmatrix} u^{t}_{11} & ... & u^{t}_{1n} \\ \vdots & \ddots & \vdots \\  u^{t}_{n1} & ... & u^{t}_{nn}\end{pmatrix}$ *
$\begin{pmatrix} l^{t}_{11} & ... & l^{t}_{1n} \\ \vdots & \ddots & \vdots \\ l^{t}_{n1} & ... & l^{t}_{nn}\end{pmatrix}$ =
$\begin{pmatrix} a^{t}_{11} & ... & a^{t}_{1n} \\ \vdots & \ddots & \vdots \\ a^{t}_{n1} & ... & a^{t}_{nn}\end{pmatrix}$

Con las submatrices principales $k~$: $A^{T}_{11} = A^{T}(1:k,1:k)$, $~L^{T}_{11} = L^{T}(1:k,1:k)~$ y $~U^{T}_{11} = U^{T}(1:k,1:k)~$ con $0 \leq k \leq n$ 

$\Rightarrow A^{T}_{11} = ~U^{T}_{11} ~ L^{T}_{11}$

Luego, como $~U^{T}_{11} ~ L^{T}_{11}~$ son invertibles, $A^{T}_{11}$ (cualquier submatriz principal de $A^{T}~$) es invertible $\Rightarrow A^{T}~$ tiene factorización $L~U$

$\Leftarrow )$ Suponemos que $A^{T}~$ tiene factorización $L~U \Rightarrow (A^{T})^{T} = A = U^{T}L^{T}$

La vuelta de la demostración es análoga a la ida, se concluye que $A$ es invertible y que como $~U^{T}_{11} ~ L^{T}_{11}~$ son inversibles, $A_{11}$ (cualquier submatriz principal de $A~$) es inversible $\Rightarrow A^{T}~$ tiene factorización $L~U$

$\therefore ~ A$ tiene factorización $L~U \Leftrightarrow A^{T}~$ tiene factorización $L~U$

A partir de esto, podemos usar que si $(I - p~W~D)^{T}$ (y sus submatrices principales) es estrictamente diagonal dominante (EDD), es decir, que para todo $i~$, $~ 0 \leq i \leq n$, $~ \sum\limits_{0 \leq j \leq n, ~ j \neq i}{(|a{ij}|)} < |a_{ii}|~ \Rightarrow (I - p~W~D)^{T}~$ es invertible (y sus submatrices principales).

$(I - p~W~D)^{T} = I - p~(W~D)^{T}$

Si $~WD = R~$ y $r_{ij} = \left\lbrace\begin{array}{c}&\ 1/c_{j} \hspace{0.5cm} &\ si ~ c_{j} \neq 0 \\&\ 0 \hspace{0.5cm} &\ si ~ c_{j} = 0 \end{array}\right. \hspace{0.3cm} $ con $\hspace{0.3cm} c_{j} = \sum\limits_{i=1}^{n}{w_{ij}}$

Entonces $(WD)^{T} = R^{T} \hspace{0.3cm} $ y $ \hspace{0.3cm} r^{~t}_{ij} = r_{ji} \Rightarrow \sum\limits^{n}_{j=1}{(r_{ij})} = 1, \hspace{0.3cm} \forall\hspace{0.1cm} 0 \leq i \leq n~$. Notemos también que $r_{ii} = 0$ para todo elemento en la diagonal de $~R~$.

Luego como $p \in (0,1) \Rightarrow p * \sum\limits^{n}_{j=1}{(r_{ij})} < 1, \hspace{0.3cm} \forall\hspace{0.1cm} 0 \leq i \leq n~$

Finalmente, si $~A = I - p~(W~D)^{T}$:

$a_{ij} = \left\lbrace\begin{array}{c}&\ 1 \hspace{0.5cm} &\ si ~ i = j \\&\ -(p/c_{j}) \hspace{0.5cm} &\ si ~ i \neq j \end{array}\right. \hspace{0.5cm}$ Notar que $c_{j}$ corresponde a la matriz $~WD~$ y no a su transpuesta.

También lo podemos reescribir como:

$a_{ij} = \left\lbrace\begin{array}{c}&\ 1 \hspace{0.5cm} &\ si ~ i = j \\&\ -(p/c_{i}) \hspace{0.5cm} &\ si ~ i \neq j \end{array}\right. \hspace{0.5cm}$ con $c_{i} = \sum\limits_{i=1}^{n}{w^{t}_{ij}}$

Entonces, como $\sum\limits_{0 \leq j \leq n, ~ j \neq i}{(|a{ij}|)} < 1~$ y $~|a_{ii}| = 1 \hspace{0.3cm} \forall\hspace{0.1cm} 0 \leq i \leq n \hspace{0.3cm} (I - p~W~D)^{T}~$ es invertible.

Más aún, como $\sum\limits_{k = 1}^{n-1}{|x_{k}|} < \sum\limits_{k = 1}^{n}{|x_{k}|}$ (la suma de números positivos es mayor a la misma suma sin uno de los sumandos).

$\Rightarrow \sum\limits_{0 \leq j \leq k, ~ j \neq i}{(|a{ij}|)} < 1~$ y $~|a_{ii}| = 1 \hspace{0.3cm} \forall\hspace{0.1cm} 0 \leq i \leq k \hspace{0.3cm}$ y todas las submatrices principales de $(I - p~W~D)^{T}~$ son EDD y, por lo tanto, invertibles.

Finalmente, como $(I - p~W~D)^{T}~$ es invertible y todas sus submatrices principales también lo son, concluimos que la matriz tiene factorización $~L~U$

Y como $(I - p~W~D)^{T}$ tiene factorización $~L~U \Leftrightarrow (I - p~W~D)~$ tiene factorización $~L~U$

$\therefore (I - p~W~D) = L~U$

##### En esta demostración, vemos que el valor de p es fundamental para probar que la matriz dada sea estrictamente diagonal dominante. De otra forma el teorema no hubiese sido valido

#### ¿Cómo influye el valor de p?

##### Si tenemos en cuenta que $p$ en el problema hace referencia a la probabilidad de que una persona siga sus links o pase a cualquier página, entonces, cuanto mayor sea $p$ (es decir, cuanto más cerca de 1 esté), menos aleatorias serán las decisiones. Por otro lado, cuanto más cerca de 0 esté $p$, más aleatorias serán las decisiones.

##### Matemáticamente hablando, podemos decir que el valor de $p$ influye de manera que, cuando este tiende a 0 la matriz está mejor condicionada. De hecho, $lim_{~p \rightarrow 0}~(I - p~W~D) = Id \hspace{0.1cm}$ por lo que su condición tiende a 1. Por el lado contrario, cuando p tiende a 1, la condición de la matriz empeora.

##### Sabiendo que al número de condición de una matriz es correcto interpretarlo como la sensibilidad a los cambios en los datos de la misma, podemos decir que las dos facetas anteriormente mencionadas de $p$ tienen mucho sentido, ya que al tender $p$ a 0, los links de la matriz pierden peso, a modo que son prácticamente irrelevantes comparado con la probabilidad de saltar a cualquier página en cualquier momento, por eso no importa cuantos links agregues o saques, la matriz se va a seguir comportando de la misma manera aleatoria. Por el lado contrario, con $p$ tendiendo a 1, los links en las páginas son mucho mas relevantes, por lo que agregar o sacar alguno de ellos podría afectar gravemente a los resultados en los rankings y scores.
